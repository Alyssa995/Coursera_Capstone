```python

import pymysql  # 导入 pymysql
import re
import time
import datetime
import requests
from lxml import etree
import string
from bs4 import BeautifulSoup
headers = {
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.106 Safari/537.36"}
```

```python
import pandas as pd
```

```python
typess=[]
names=[]
locations=[]
times=[]
urls=[]
des=[]
res=[]
```

```python
for i in range(1,3):
    final_url = 'http://careerhx.zhiye.com/social/?PageIndex='+str(i)
    response = requests.get(final_url,headers=headers)
    r = response.text
    html = etree.HTML(r,etree.HTMLParser())
    r0 = html.xpath('//div[@class="shjc"]//p')
    r1 = html.xpath('//div[@class="shjc"]//dd')
    r2 = html.xpath('//div[@class="shjc"]//i')
    r3 = html.xpath('//div[@class="shjc"]//em')
    r4 = html.xpath('//div[@class="shjc"]//span/a/@href')
    for j in range(0,10):
        name = r0[j].xpath('string(.)')
        types = r1[j].xpath('string(.)')
        location = r2[j].xpath('string(.)')
        time = r3[j].xpath('string(.)')
        url='http://careerhx.zhiye.com'+r4[j]
        response2 = requests.get(url,headers=headers)
        r_1 = response2.text
        html2 = etree.HTML(r_1,etree.HTMLParser())
        res=html2.xpath('//div[@class="shjg"]//p')
        res1=res[0].xpath('string(.)')
        names.append(name)
        typess.append(types)
        locations.append(location)
        times.append(time)
        urls.append(url)
        des.append(res1)
```

```python
for i in range(3,4):
    final_url = 'http://careerhx.zhiye.com/social/?PageIndex='+str(i)
    response = requests.get(final_url,headers=headers)
    r = response.text
    html = etree.HTML(r,etree.HTMLParser())
    r0 = html.xpath('//div[@class="shjc"]//p')
    r1 = html.xpath('//div[@class="shjc"]//dd')
    r2 = html.xpath('//div[@class="shjc"]//i')
    r3 = html.xpath('//div[@class="shjc"]//em')
    r4 = html.xpath('//div[@class="shjc"]//span/a/@href')
    for j in range(0,6):
        name = r0[j].xpath('string(.)')
        types = r1[j].xpath('string(.)')
        location = r2[j].xpath('string(.)')
        time = r3[j].xpath('string(.)')
        url='http://careerhx.zhiye.com'+r4[j]
        response2 = requests.get(url,headers=headers)
        r_1 = response2.text
        html2 = etree.HTML(r_1,etree.HTMLParser())
        res=html2.xpath('//div[@class="shjg"]//p')
        res1=res[0].xpath('string(.)')
        names.append(name)
        typess.append(types)
        locations.append(location)
        times.append(time)
        urls.append(url)
        des.append(res1)
```

```python
jobs=pd.DataFrame(names,columns = ['职位名称'])
jobs['职位类型']=typess
jobs['工作地点']=locations
jobs['发布时间']=times
jobs['申请职位']=url
jobs['职位描述']=des
jobs
```
